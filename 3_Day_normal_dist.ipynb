# Normal Dist

***effect of mean ***

```python
values=np.random.normal(loc=0,scale=1,size=10000)
plt.hist(values,bins=20,density=True,color='black')

x=np.linspace(values.min(),values.max(),100)
from scipy.stats import norm
p = norm.pdf(x, loc=np.mean(values), scale=np.std(values))

plt.plot(x,p)

# change mean
values=np.random.normal(loc=1,scale=1,size=10000)
plt.hist(values,bins=20,density=True)

x=np.linspace(values.min(),values.max(),100)
from scipy.stats import norm
p = norm.pdf(x, loc=np.mean(values), scale=np.std(values))

plt.plot(x,p)

# centre is change when mean change
```

***Effect of Std***

```python
values=np.random.normal(loc=0,scale=1,size=10000)
plt.hist(values,bins=20,density=True,color='black')

x=np.linspace(values.min(),values.max(),100)
from scipy.stats import norm
p = norm.pdf(x, loc=np.mean(values), scale=np.std(values))

plt.plot(x,p)

# change std
values=np.random.normal(loc=0,scale=1.7,size=10000)
plt.hist(values,bins=20,density=True)

x=np.linspace(values.min(),values.max(),100)
from scipy.stats import norm
p = norm.pdf(x, loc=np.mean(values), scale=np.std(values))

plt.plot(x,p)

# when change std peak height will changes
```

# parameteric Density Estimation

```python
sample=np.random.normal(loc=50,scale=5,size=100)
sample.mean()
```

```python
plt.hist(sample,bins=20)
```

```python
# calculate mean and std
sample_mean=sample.mean()
sample_std=sample.std()
```

```python
from scipy.stats import norm
dist=norm(sample_mean,sample_std)
```

```python
values=np.linspace(sample.min(),sample.max(),100)
probabilities=[dist.pdf(value) for value in values]
```

```python
values.mean()
```

```python
# plot the histogram and pdf
plt.hist(sample,bins=10,density=True)
plt.plot(values,probabilities)
```

```python
import seaborn as sns
sns.distplot(sample)
```

# KDE

```python
# generate data
from numpy.random import normal
sample1=normal(loc=20,scale=5,size=300)
sample2=normal(loc=50,scale=5,size=700)
sample=np.hstack((sample1,sample2))
sample
```

```python
plt.hist(sample,kind='bar',bins=20)
```

```python
from sklearn.neighbors import KernelDensity

model = KernelDensity(bandwidth=5, kernel='gaussian')

# convert data to a 2D array
sample = sample.reshape((len(sample), 1))

model.fit(sample)
```

```python
values = np.linspace(sample.min(),sample.max(),100)
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = np.exp(probabilities)
plt.hist(sample, bins=50, density=True)
plt.plot(values[:], probabilities)
plt.show()
```

```python
sns.kdeplot(sample.reshape(1000),bw_adjust=2)
```
