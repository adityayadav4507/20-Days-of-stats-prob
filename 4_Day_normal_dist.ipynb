```python
titanic=pd.read_csv('/content/titanic.csv')
titanic.head()
```

```python
sns.kdeplot(data=titanic,x='Age',hue='Survived')
```

```python
titanic['Age'].mean()
```

```python
X=(titanic['Age']-titanic['Age'].mean())/(titanic['Age'].std())
```

```python
sns.kdeplot(X) #  std normal dist
```

```python
# 2 D density plot

sns.jointplot(data=titanic,x='Age',y='Fare',kind='kde',fill=True,cbar=True)
```
