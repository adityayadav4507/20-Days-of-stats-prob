```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/titanic.csv')
df.head()
```

```python
# count passanger in each class

count_class=df['Pclass'].value_counts().sort_index()
count_class
```

```python
# calculate expected count

expected_count=len(df['Pclass'])/3
expected_count
```

```python
# perform chi-square goodness of test

chi2 , p_value=stats.chisquare(count_class,expected_count)

print("chi-square statistic:",chi2)
print("p-value:",p_value)
```

# test for independance

```python
# perform chi-square test for independance

from scipy.stats import chi2_contingency

# create a contingency table

contingency_table=pd.crosstab(df['Survived'],df['Pclass'])
contingency_table
```

```python
# perform the chi-square test

chi2,p_value,dof,expected=chi2_contingency(contingency_table)

print("chi-square statistic:",chi2)
print("p-value:",p_value)
print("degree of freedom:",dof)
print("expected frequency table:")
print(expected)
```

```python
alpha=0.05
if alpha>p_value:
  print("reject the null hypothesis")
else:
  print("accept the null hypothesis")
```
