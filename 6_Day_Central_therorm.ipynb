# central theorm

# 1. uniform dist

```python
# set parameters
num_samples=1000
sample_size=30
distrbution_range=(0,1)

# generate samples form  uniform distribution
samples=np.random.uniform(distrbution_range[0],distrbution_range[1],(num_samples,sample_size)) # each set =30 , no.of set 1000

plt.hist(samples)
```

```python
# sample means
sample_mean=np.mean(samples,axis=1)
plt.hist(sample_mean,bins=30,density=True)
plt.xlabel('sample mean')
plt.ylabel('density')
plt.show()
```

# 2. expo dist

```python
# set parameters
num_samples=1000
sample_size=30
lambda_sample=2

# generate samples form  expo
samples=np.random.exponential(lambda_sample,(num_samples,sample_size)) # each set =30 , no.of set 1000

plt.hist(samples)
```

```python
# mean of samples
sample_mean=np.mean(samples,axis=1)
plt.hist(sample_mean,bins=30,density=True)
plt.xlabel('sample mean')
plt.ylabel('density')
plt.show()
```

# 3. poisson dist
# 4. gamma dist
# 5. bionomial dist

```python
# set parameters

num_samples=1000
sample_size=30

# 3. poision dist parametrs
poisson_lambda=5
# 4. gamma dist parameters
gamma_shape=2
gamma_scale=1
# 5 . bionomial dist
n_binomial=10
p_binomial=0.5

# Generate samples from the dist
samples_poisson=np.random.poisson(poisson_lambda,(num_samples,sample_size))
samples_gamma=np.random.gamma(gamma_shape,gamma_scale,(num_samples,sample_size))
samples_binomial=np.random.binomial(n_binomial,p_binomial,(num_samples,sample_size))

# calculate means
poisson_mean=np.mean(samples_poisson,axis=1)
gamma_mean=np.mean(samples_gamma,axis=1)
binomial_mean=np.mean(samples_binomial,axis=1)

# plot histogram of samples mean

fig ,axs = plt.subplots(1,3,figsize=(15,5))

# 3. poisson dist
axs[0].hist(poisson_mean,bins=30,density=True)
axs[0].set_xlabel('sample mean')
axs[0].set_ylabel('density')
axs[0].set_title('poisson distribution')

# 4. gamma dist
axs[1].hist(gamma_mean,bins=30,density=True)
axs[1].set_xlabel('sample mean')
axs[1].set_ylabel('density')
axs[1].set_title('gamma distribution')

# 5. bionomial dist
axs[2].hist(binomial_mean,bins=30,density=True)
axs[2].set_xlabel('sample mean')
axs[2].set_ylabel('density')
axs[2].set_title('binomial distribution')
```

# mean of sample = mean of population

```python
# set parametrs
num_samples=10000
sample_size=50
gamma_shape=2
gamma_scale=1


# mean of population
population_mean=gamma_scale*gamma_shape
population_var=gamma_shape*gamma_scale**2

# generate samples from gamma distribution
samples=np.random.gamma(gamma_shape,gamma_scale,(num_samples,sample_size))

samples_means=np.mean(samples,axis=1)

# mean of sample

sample_mean=np.mean(samples_means)
sample_var=np.var(samples_means)


print(f"population mean : {population_mean}")
print(f"population variance : {population_var}")
print()
print(f"sample mean : {sample_mean}")
print(f"sample variance : {sample_var}")

```
